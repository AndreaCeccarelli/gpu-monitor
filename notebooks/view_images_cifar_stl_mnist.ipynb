```python
import numpy as np
from PIL import Image
from skimage import data
from skimage.color import rgb2gray
from matplotlib.pyplot import imshow

IMAGE=5 #any number (mnist, cifar up to 10000, stl up to 8000)
```

## Show image for stl

```python
def to_gray_uint(image):
    return np.uint8(rgb2gray(image) * 255)
```

```python
data1=np.load("/home/andrea/gpu-monitor/fullattacks/DeepFool_stl10_pytorch.npy")
data = np.swapaxes(data1, 1, 3).astype(np.float32)
data = np.swapaxes(data, 1, 2).astype(np.float32)
#cifar:
#'plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck'
#stl:
#airplane, bird, car, cat, deer, dog, horse, monkey, ship, truck
```

```python
d = data[IMAGE].astype(np.float32) #/ 1 # normalize the data to 0 - 1
d = 255 * d # Now scale by 255
img = d.astype(np.uint8)
#print(data[8])
imshow(np.asarray(img))
```
